```python
from sklearn.pipeline import Pipeline
from feature_engine.datetime import DatetimeFeatures
from feature_engine.selection import DropConstantFeatures

pipe = Pipeline([
    ('datetime_extraction', DatetimeFeatures(
        variables="date_time",
        features_to_extract=["year", "day_of_month", "minute", "second"],
        drop_original=False # Orijinal tarihi korumak istersen
    )),
    ('drop_constants', DropConstantFeatures()) # Gereksiz sabitleri siler
])

data_transf = pipe.fit_transform(data)
```
