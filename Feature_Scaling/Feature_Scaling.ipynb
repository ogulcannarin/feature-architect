Feature Scaling (Özellik Ölçeklendirme) Nedir?
En Kısa Tanım: Veri setindeki farklı ölçü birimlerine sahip sayısal sütunları (metrekare, oda sayısı, fiyat vb.), birbirleriyle kıyaslanabilir hale getirmek için aynı sayısal aralığa çekme işlemidir.

1. Neden Yapıyoruz? (Temel Sebep)
Makine öğrenmesi modelleri sadece "sayılara" bakar, birimin ne olduğunu bilmez.

Problem: Evin metrekaresi 2500 iken banyo sayısı 2'dir.

Modelin Hatası: Model, 2500 sayısının 2'den çok daha büyük olduğunu görüp "Metrekare bin kat daha önemlidir" diye yanlış bir genelleme yapabilir.

Çözüm: Scaling ile her iki değeri de (örneğin) 0 ile 1 arasına getiririz. Böylece model, her iki özelliğe de adil (eşit) davranır.

2. En Çok Kullanılan İki Yöntem
A) Standardization (Standardizasyon / StandardScaler)
Veriyi ortalaması 0, standart sapması 1 olacak şekilde dönüştürür.

Mantık: Her bir değerden ortalamayı çıkarır ve standart sapmaya böler.

Sonuç: Çoğu değer -3 ile +3 arasında toplanır.

Ne zaman kullanılır? Veride aykırı değerler (outliers) varsa daha güvenlidir.

B) Normalization (Normalizasyon / MinMaxScaler)
Veriyi tam olarak 0 ile 1 (veya -1 ile 1) arasına sıkıştırır.

Mantık: (Değer - Min) / (Max - Min)

Sonuç: En küçük değer 0, en büyük değer 1 olur.

Ne zaman kullanılır? Verinin dağılımı hakkında bir bilgin yoksa veya derin öğrenme (Deep Learning) yapıyorsan.

3. Scaling Yapmazsak Ne Olur? (Kritik Uyarı!)
Özellikle şu modellerdeScaling şarttır:

Mesafe Temelli Modeller: KNN, K-Means (Mesafe ölçtükleri için büyük sayılar mesafeyi domine eder).

Gradyan Temelli Modeller: Lineer Regresyon, Lojistik Regresyon, Yapay Sinir Ağları (Scaling olmazsa modelin "öğrenmesi" çok uzun sürer veya hiç bitmez).

Not: Ağaç temelli modeller (Random Forest, XGBoost) genellikle Scaling gerektirmez ama yapmak da bir zarar vermez.

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# 1. Örnek bir mini veri seti oluşturalım
data = {
    'GrLivArea': [1500, 2500, 800, 3500],
    'FullBath': [2, 3, 1, 4]
}
demo_df = pd.DataFrame(data)
```

```python
scaler = StandardScaler()
demo_df_scaled = pd.DataFrame(scaler.fit_transform(demo_df), columns=demo_df.columns)
```
